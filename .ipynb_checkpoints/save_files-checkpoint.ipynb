```python
import shutil
import fnmatch
#import pyFAI

import os

os.getcwd()
```

```python
def get_all_filepaths(root_path, ext):

    import os
    all_files = []
    nother = 1
    nagb = 1
    ngisaxs = 1
    ngiwaxs = 1
    nwaxs = 1
    nsaxs = 1
    nauto = 1
    nbeam = 1
    
    for root, dirs, files in os.walk(root_path):
        for filename in files:
            if filename.lower().endswith(ext):
                all_files.append(os.path.join(root, filename))
                # filename_new = filename[:-4]

                if fnmatch.fnmatch(os.path.join(root, filename).lower(),"*gisaxs*"):
                    print(os.path.join(root, filename))
                    zahl = "000000" + str(ngisaxs)
                    file_new = "gisaxs" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/GISAXS', file_new))
                    ngisaxs = ngisaxs + 1
                elif fnmatch.fnmatch(os.path.join(root, filename).lower(),"*giwaxs*"):
                    #print os.path.join(root, filename)
                    zahl = "000000" + str(ngiwaxs)
                    file_new = "giwaxs" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/GIWAXS', file_new))
                    ngiwaxs = ngiwaxs + 1
                
                elif fnmatch.fnmatch(os.path.join(root, filename).lower(),"*saxs*"):
                    #print os.path.join(root, filename)
                    zahl = "000000" + str(nsaxs)
                    file_new = "saxs" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/SAXS', file_new))
                    nsaxs = nsaxs + 1
                    
                elif fnmatch.fnmatch(os.path.join(root, filename).lower(),"*waxs*"):
                    #print os.path.join(root, filename)
                    zahl = "000000" + str(nwaxs)
                    file_new = "waxs" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/WAXS', file_new))
                    nwaxs = nwaxs + 1
                    
                elif filename.lower().startswith("agb"):
                    zahl = "000000" + str(nagb)
                    file_new = "AgB" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/AgB', file_new))
                    nagb = nagb + 1
                    
                elif filename.lower().startswith("auto"):
                    zahl = "000000" + str(nauto)
                    file_new = "auto" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/auto', file_new))
                    nauto = nauto + 1
                elif filename.lower().startswith("beamstop"):
                    zahl = "000000" + str(nbeam)
                    file_new = "beam" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename), os.path.join('Learning/beamstop', file_new))
                    nbeam = nbeam + 1
                else:
                    zahl = "000000" + str(nother)
                    file_new = "other" + zahl.zfill(5) + ".edf"
                    shutil.copy(os.path.join(root, filename),os.path.join("Learning/other", file_new))
                    nother = nother + 1
                    #shutil.copy(os.path.join(root, filename),os.path.join("Learning\Transmission", file_new))
                    ##check path
                    #print(os.path.join(root, filename))
                    #print(os.path.join("Learning/Transmission", file_new))
                    
                    
    return all_files
# Example: print all the .py files in './python'
path = os.getcwd()
a = get_all_filepaths(path, "edf")
#b = pyFAI.load("AgB.edf")
#print b
#print os.path.join(root, filename)
```
